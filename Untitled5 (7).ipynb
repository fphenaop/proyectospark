```python
#Definimos el  esquema para los datos, con el fin optimizar en el momento que se lean los datos, evitar 'inferschema'
airbnb_schema = StructType([
    StructField("Title",StringType(),True),
    StructField("Detail",StringType(),True),
    StructField("Date",StringType(),True),
    StructField("Price(in dollar)",DoubleType (),True),
    StructField("Offer price(in dollar)",DoubleType (),True),
    StructField("Review and rating",StringType(),True),
    StructField("Number of bed",StringType(),True)
])

airbnb_luxe_schema = StructType([
    StructField("Luxe name",StringType(),True),
    StructField("Date",StringType(),True),
    StructField("Price(in dollar)",DecimalType(12, 2),True),
    StructField("Distance",StringType(),True),
])
```

```python
#Top10 ciudades más costosas en promedio para alquiler

top_avg_cost = ss.sql("""
       SELECT City,
           AVG(`Price(in dollar)`) AS Avg_Price
        FROM tempAirbnView
        GROUP BY City
        ORDER BY Avg_Price DESC
        LIMIT 10
""")

top_avg_cost.show(truncate=False)
```

```python
#Top 10 de las propiedades con mayor numero de camas en COUNTRY US por CITY
ranked_properties_df = ss.sql("""
    WITH RankedProperties AS (
        SELECT `City`,
               `Title`,
               `Number of beds`,
               `Country`,
               ROW_NUMBER() OVER (PARTITION BY `City` ORDER BY CAST(`Number of beds` AS INT) DESC) AS rn
        FROM tempAirbnView
        WHERE Country = 'US'
          AND `City` IS NOT NULL
          AND `Number of beds` IS NOT NULL
    )
    SELECT `City`,
           `Title`,
           `Number of beds`
    FROM RankedProperties
    WHERE rn = 1
    ORDER BY CAST(`Number of beds` AS INT) DESC
    LIMIT 10
""")

ranked_properties_df.show(truncate=False)
```

```python
#media por city de rating menor a 4  y Price(in dollar) de manera descensente
average_ratings_prices_df = ss.sql("""
    SELECT
        City,
        AVG(Rating) AS Avg_Rating,
        AVG(`Price(in dollar)`) AS Avg_Price
    FROM
        tempAirbnView
    GROUP BY
        City
    HAVING
        AVG(Rating) < 4
    ORDER BY
        Avg_Rating DESC,
        Avg_Price DESC
""")

average_ratings_prices_df.show(truncate=False)
```

```python
#hospedeja mas costoso en hospedaje normal y luxury
max_price = (airbnbDF
            .agg(fct.max("Price(in dollar)").alias("max_price_normal"),fct.min("Price(in dollar)").alias("min_price_normal"),fct.mean("Price(in dollar)").alias("avg_price_normal"))
            .join(airbnb_luxeDF
                .agg(fct.max("Price(in dollar)").alias("max_price_luxe"),fct.min("Price(in dollar)").alias("min_price_luxury"),fct.mean("Price(in dollar)").alias("avg_price_luxury") )
            )
)
max_price.show(truncate=False)
```

```python
#Revisando si las ciudades con mejores puntuaciones en hospedaje normal  tienen hospedaje luxury

rating_lodgment = (
    airbnbDF.groupBy("City")
    .agg(fct.avg("Rating").alias("avg_rating"))
    .join(
        airbnb_luxeDF.select("City").distinct(),  # Selecciona y elimina duplicados de "City" en airbnb_luxeDF
        ["City"], "left"
    )
    .withColumn("has_luxury", fct.when(fct.col("City").isin(airbnb_luxeDF.select("City").distinct().rdd.flatMap(lambda x: x).collect()), 1).otherwise(0))
    .filter(fct.col("has_luxury") == 1)
    .select("City", "avg_rating", "has_luxury")
    .orderBy(fct.col("avg_rating").desc())
)

rating_lodgment.show(100, False)
```

```python

# Función para tokenizar el texto en palabras, manejando nulos
def tokenize(text):
    if text is None:
        return []  # Devuelve un array vacío si el texto es nulo
    return text.lower().split()

# Registrar la UDF en PySpark
tokenize_udf = fct.udf(tokenize, ArrayType(StringType()))

# Reemplazar valores nulos en la columna Detail con una cadena vacía
airbnbDF_non_null = airbnbDF.fillna({'Detail': ''})

# Aplicar la UDF para tokenizar el texto en el campo Detail
airbnbDF_tokenized = airbnbDF_non_null.withColumn("tokens", tokenize_udf(fct.col("Detail")))

# Explode para convertir los arrays en filas individuales
airbnbDF_words = airbnbDF_tokenized.select(fct.explode(fct.col("tokens")).alias("word"))

# Contar la frecuencia de cada palabra
word_counts = airbnbDF_words.groupBy("word").count().orderBy(fct.col("count").desc())

# Convertir a Pandas para graficar
word_counts_pd = word_counts.limit(20).toPandas()  # Limitar a las 20 palabras más comunes

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x="count", y="word", data=word_counts_pd, palette="viridis")
plt.title("Top 20 Palabras más Comunes en el campo 'Detail'")
plt.xlabel("Frecuencia")
plt.ylabel("Palabra")
plt.show()
```
